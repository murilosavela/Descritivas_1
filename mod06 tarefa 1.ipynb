#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
pd = pd.read_csv('sinasc_RO_2019.csv')
pd.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
fig, axes = plt.subplots(2,2)
plt.subplots_adjust(wspace=0.4, hspace=0.5)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=pd)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=pd)
sns.countplot(ax = axes[1, 0], x='SEXO', data=pd)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=pd)
```

```python
pd['GESTACAO'].value_counts().sort_index().plot.bar()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x='APGAR5', y='CONSULTAS', data=pd)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=pd)
gestacao = sns.countplot(ax = axes[1, 0], x='GESTACAO', data=pd)
sns.countplot(ax = axes[0, 1], x='ESTCIVMAE', data=pd)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=pd)

gestacao.set_xticklabels(gestacao.get_xticklabels(), rotation=90, horizontalalignment='right')

plt.subplots_adjust(wspace=0.2, hspace=0.2)
```
